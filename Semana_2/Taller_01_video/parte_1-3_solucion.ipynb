### Ejercicio 1.3 — Listas

```python
casos = [120, 45, 300, 15, 210]

print("Primer elemento:", casos[0])
print("Último elemento:", casos[-1])
print("Longitud:", len(casos))
print("Suma total:", sum(casos))
print("Mínimo:", min(casos))
print("Máximo:", max(casos))

casos.append(500)
casos.sort(reverse=True)
print("Lista final:", casos)
```

**Pregunta de reflexión:** si la lista tiene 6 elementos después del `append`, ¿qué pasaría si intentas acceder a `casos[6]`? ¿Por qué?

**Respuesta:** produciría un `IndexError`. Las listas están indexadas desde 0, así que una lista de 6 elementos tiene posiciones válidas de `casos[0]` a `casos[5]`. `casos[6]` estaría pidiendo un séptimo elemento que no existe.

### Ejercicio 1.5 — Funciones y control de flujo

```python
def clasificar_riesgo(casos):
    if casos < 100:
        return "bajo"
    elif casos <= 300:
        return "medio"
    else:
        return "alto"

print(clasificar_riesgo(50))
print(clasificar_riesgo(150))
print(clasificar_riesgo(400))
```

## Parte 2 — Pandas aplicado a un dataset tipo COVID-19

### Preparación del dataset (ejecutar antes de la Parte 2)

```python
import pandas as pd

data = pd.read_csv("covid_19_data.csv", parse_dates=["ObservationDate", "Last Update"])
```

### Ejercicio 2.1 — Exploración inicial

```python
print(data.head())
print(data.tail())
print(data.shape)
```

### Ejercicio 2.2 — Información y tipos de dato

```python
data.info()
```

**Pregunta de reflexión:** si `ObservationDate` apareciera como `object` (texto), ¿qué parámetro de `pd.read_csv` usarías para corregirlo?

**Respuesta:** el parámetro `parse_dates`, indicando el nombre de la(s) columna(s) de fecha, por ejemplo `pd.read_csv("covid_19_data.csv", parse_dates=["ObservationDate"])`.

### Ejercicio 2.3 — Estadísticas básicas

```python
data.describe()
```

### Ejercicio 2.4 — Valores únicos

```python
print(data["Country/Region"].unique())
```

### Ejercicio 2.5 — Filtrado de filas

```python
mexico = data[data["Country/Region"] == "Mexico"]
print(mexico["Country/Region"].unique())
mexico.head()
```

### Ejercicio 2.6 — Última fecha disponible

```python
print(data["ObservationDate"].max())
```

### Ejercicio 2.7 — Agrupar y agregar (`groupby`)

```python
casos_max_pais = data.groupby("Country/Region")["Confirmed"].max().reset_index()
casos_max_pais
```

### Ejercicio 2.8 — Ordenar resultados

```python
casos_max_pais = casos_max_pais.sort_values("Confirmed", ascending=False)
casos_max_pais
```

### Ejercicio 2.9 — Visualización de barras

```python
import plotly.express as px

fig = px.bar(
    casos_max_pais,
    x="Country/Region", y="Confirmed",
    title="Casos confirmados máximos por país"
)
fig.show()
```

### Ejercicio 2.10 — Serie de tiempo

```python
fig = px.line(
    mexico,
    x="ObservationDate", y="Confirmed",
    title="Evolución de casos confirmados en México"
)
fig.show()
```

### Ejercicio 2.11 — Tabla dinámica (`pivot_table`)

```python
resumen = data.pivot_table(
    index=["Country/Region", "Province/State"],
    values=["Confirmed", "Deaths", "Recovered"],
    aggfunc="sum"
)
resumen
```

**Reto adicional:** filtrar solo `"Mainland China"` usando `.loc`.

```python
resumen.loc["Mainland China"]
```

### Ejercicio 2.12 — Muertes totales por fecha

```python
muertes_por_fecha = data.groupby("ObservationDate")["Deaths"].sum().reset_index()

fig = px.line(
    muertes_por_fecha,
    x="ObservationDate", y="Deaths",
    title="Total de muertes por fecha de observación (todos los países)"
)
fig.show()
```

## Parte 3 — Reto integrador (opcional)

```python
us = data[data["Country/Region"] == "US"]
us["mes"] = us["ObservationDate"].dt.month

promedio_recuperados_mes = us.groupby("mes")["Recovered"].mean().reset_index()
promedio_recuperados_mes = promedio_recuperados_mes.sort_values("Recovered", ascending=False)
promedio_recuperados_mes
```

```python
fig = px.bar(
    promedio_recuperados_mes,
    x="mes", y="Recovered",
    title="Promedio de recuperados acumulados por mes — US"
)
fig.show()
```

**Pregunta de cierre:** ¿qué tipo de análisis (descriptivo, predictivo o prescriptivo) corresponde a todo lo que hicimos en este ejercicio? ¿Qué haría falta para llevarlo al siguiente nivel de análisis?

**Respuesta:** todo lo realizado en este ejercicio es **análisis descriptivo**: describe qué pasó (casos máximos por país, evolución histórica, muertes acumuladas, promedios mensuales), pero no proyecta valores futuros ni recomienda acciones. Para llevarlo al siguiente nivel (**predictivo**) haría falta construir un modelo — por ejemplo una regresión lineal o un modelo de series de tiempo — que use los datos históricos para proyectar cómo evolucionarán los casos, muertes o recuperados en fechas futuras. Y para llegar al nivel **prescriptivo**, ya con las proyecciones predictivas, habría que traducir esos resultados en recomendaciones concretas de acción (p. ej. medidas de confinamiento según el crecimiento proyectado de casos).
