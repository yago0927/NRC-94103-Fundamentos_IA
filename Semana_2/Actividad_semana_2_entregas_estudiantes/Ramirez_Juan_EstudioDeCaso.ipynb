```python
# ==================================================
# RELACIÓN ENTRE EL CURSO DE PREPARACIÓN Y SUS PUNTAJES
# ==================================================
```

```python
# ==================================================
# PROMEDIO DE PUNTAJES SEGUN CURSO DE PREPARACON
# ==================================================

promedios_curso = df.groupby("test preparation course")[
    ["math score", "reading score", "writing score"]
    ].mean()

promedios_curso
```

```python
# ==================================================
# PROMEDIO DE PUNTAJES SEGUN GENERO
# ==================================================

promedios_genero = df.groupby("gender")[
    ["math score", "reading score", "writing score"]
].mean()

promedios_genero
```

# ==================================================
# PROMEDIO DE PUNTAJES SEGUN NIVEL EDUCATIVO DE PADRES
# ==================================================

```python
promedios_padres = df.groupby("parental level of education")[
    ["math score", "reading score", "writing score"]
].mean()

promedios_padres
```
