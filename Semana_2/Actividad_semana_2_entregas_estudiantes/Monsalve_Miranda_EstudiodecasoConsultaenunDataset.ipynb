```python
# a. Valores nulos por columna
print("--- a. Valores nulos por columna ---")
print(df.isnull().sum())
```

```python
# b. Valores duplicados
print("\n--- b. Cantidad de filas duplicadas ---")
duplicados = df.duplicated().sum()
print(f"Total de filas duplicadas: {duplicados}")
```

```python
# c. Valores atípicos (en la variable 'math score')
# Usaremos el método del Rango Intercuartílico (IQR) para detectar notas atípicamente bajas o altas
print("\n--- c. Detección de valores atípicos en 'math score' ---")
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar los estudiantes que están fuera de los límites
atipicos = df[(df['math score'] < limite_inferior) | (df['math score'] > limite_superior)]

print(f"Límite inferior para notas normales: {limite_inferior}")
print(f"Límite superior para notas normales: {limite_superior}")
print(f"Cantidad de valores atípicos detectados: {len(atipicos)}")
if len(atipicos) > 0:
    print("\nEjemplo de registros con valores atípicos:")
    print(atipicos[['gender', 'math score']].head())
```

```python
# a. Variables Categóricas (Frecuencias y Porcentajes)
# ==========================================
print("--- a. Análisis de Variables Categóricas ---")

# Variable 1: Género (gender)
print("\n--- Frecuencia de Género ('gender') ---")
print(df["gender"].value_counts())
print("\n--- Porcentaje de Género ('gender') ---")
print(df["gender"].value_counts(normalize=True) * 100)

# Variable 2: Nivel de Educación de los Padres (parental level of education)
print("\n--- Frecuencia de Educación de los Padres ---")
print(df["parental level of education"].value_counts())
print("\n--- Porcentaje de Educación de los Padres ---")
print(df["parental level of education"].value_counts(normalize=True) * 100)
```

```python
# b. Variables Numéricas (Medidas Estadísticas)
# ==========================================
print("\n==========================================\n")
print("--- b. Análisis de Variables Numéricas ---")

# Usamos describe() para obtener media, mediana (50%), min, max y desviación estándar automáticamente

# Variable 1: Calificación de Matemáticas (math score)
print("\n--- Estadísticas de Matemáticas ('math score') ---")
print(df["math score"].describe())

# Variable 2: Calificación de Lectura (reading score)
print("\n--- Estadísticas de Lectura ('reading score') ---")
print(df["reading score"].describe())
```
