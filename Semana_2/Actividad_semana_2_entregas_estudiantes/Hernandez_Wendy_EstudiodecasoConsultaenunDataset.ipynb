```python
# ==============================================================================
# 1. IMPORTACIÓN DE LIBRERÍAS Y CARGA DEL DATASET
# ==============================================================================
import pandas as pd
import numpy as np

# Carga del dataset desde el archivo local subido a Colab
df = pd.read_csv('StudentsPerformance.csv')

print("--- Visualización Inicial del Dataset ---")
print(df.head())

# ==============================================================================
# 2. COMPRENSIÓN DE VARIABLES (TAMAÑO, NOMBRES Y TIPOS)
# ==============================================================================
print("\n--- a. Tamaño del Dataset (Filas, Columnas) ---")
print(df.shape)

print("\n--- b. Nombres de las Columnas ---")
print(df.columns.tolist())

print("\n--- c. Tipos de Datos ---")
print(df.dtypes)

print("\n--- d. Resumen General ---")
print(df.info())

# ==============================================================================
# 3. DIAGNÓSTICO DE CALIDAD DE DATOS
# ==============================================================================
print("\n--- a. Valores Nulos por Columna ---")
print(df.isnull().sum())

print("\n--- b. Registros Duplicados ---")
print(f"Total de registros duplicados: {df.duplicated().sum()}")

print("\n--- c. Valores Atípicos (Outliers) en Variable Numérica ---")
# Analizando la variable 'math score' mediante los cuartiles (Regla del IQR)
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers = df[(df['math score'] < limite_inferior) | (df['math score'] > limite_superior)]
print(f"Límites para math score: {limite_inferior} a {limite_superior}")
print(f"Cantidad de valores atípicos detectados en math score: {len(outliers)}")

# ==============================================================================
# 4. ANÁLISIS EXPLORATORIO INICIAL (EDA)
# ==============================================================================
print("\n--- a. Variables Categóricas (Frecuencias y Porcentajes) ---")
for col in ['gender', 'race/ethnicity']:
    print(f"\nDistribución para: {col}")
    print(df[col].value_counts())
    print(df[col].value_counts(normalize=True) * 100)

print("\n--- b. Variables Numéricas (Estadísticas Descriptivas) ---")
print(df[['math score', 'reading score', 'writing score']].describe())

# ==============================================================================
# 5. RELACIÓN ENTRE VARIABLES (Para responder la pregunta D)
# ==============================================================================
print("\n--- Relación: Promedio de calificaciones por nivel de educación de los padres ---")
print(df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean())
```
