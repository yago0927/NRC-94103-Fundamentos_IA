## Comprensión del conjunto de datos

El dataset contiene información sobre el rendimiento académico de estudiantes en pruebas de matematicas, lectura y escritura, adémas de varibales relacionadas con su contexto familiar y educativo.

Las variables se calsifican en dos grupos:

### Variables categóricas

*   gender

*   parental level of education

Estas representan categorías y permiten analizar frecuencias, porcentajes y distribuciones.

### Variables numéricas

*   math score
*   reading score
*   writing score

Estas representan calificaciones obtenidas y permiten calcular medidas estadísticas como la media, mediana y desviación estándar.

```python
# Verificar valores nulos
df.isnull().sum()
```

## Valores nulos

Se verifica la existencia de valores nulos mediante la función `isnull().sum()`. Los resultados muestran que ninguna de las variables presenta datos vacios, por lo que no fue necesario aplicar técnicas de imputación o eliminación de registros.

```python
# Buscar registros duplicados
df.duplicated().sum()
```

## Registros duplicados

Se revisa la existencia de registros duplicados utilizando la función `duplicated().sum()`. El resultado obtenido indica que el conjunto de datos no contiene registros repetidos, por lo que no fue necesario realizar procesos de depuración.

```python
# Boxplot para identificar valores atípicos en la variable math score
plt.figure(figsize=(8,5))
plt.boxplot(df["math score"])
plt.title("Boxplot de Math Score")
plt.ylabel("Puntaje")
plt.show()
```

Un **Boxpot** es un gráfico que permite identificar la mediana, el rango donde está la mayoría de los datos y valores atípicos. Visualmente se ve una caja con una línea en el centro.

En caso de observar valores fuera de los límites del diagrama, estos pueden considerarse valores atípicos. Sin embargo, al tratarse de calificaciones reales obtenidas por estudiantes, dichos valores representan observaciones válidas y no necesariamente errores en el conjunto de datos.

## Análisis Variable Reading Score

La variable **Reading Score** corresponde a la calificación obtenida por los estudiantes en la prueba de lectura. El análisis descriptivo permite conocer el promedio, la variabilidad y la distribución general de los resultados obtenidos.

## Histogramas

```python
# Math Score
plt.figure(figsize=(6,4))
plt.hist(df["math score"], bins=20, edgecolor="black")
plt.title("Distribución de Math Score")
plt.xlabel("Puntaje")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Reading Score
plt.figure(figsize=(6,4))
plt.hist(df["reading score"], bins=20, edgecolor="black")
plt.title("Distribución de Reading Score")
plt.xlabel("Puntaje")
plt.ylabel("Frecuencia")
plt.show()
```

## Relaciones entre variables

```python
# Relación 1
plt.figure(figsize=(6,5))
plt.scatter(df["math score"], df["reading score"])
plt.title("Math Score vs Reading Score")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
```

```python
# Relación 2
plt.figure(figsize=(6,5))
plt.scatter(df["reading score"], df["writing score"])
plt.title("Reading Score vs Writing Score")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.show()
```

Se logra observar una relación positiva entre las calificaciones de lectura y escritura.
