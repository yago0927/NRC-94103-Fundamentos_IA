```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
file_path = "StudentsPerformance.csv"
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "spscientist/students-performance-in-exams",
  file_path,)
```

```python
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
iqr = Q3 - Q1
inf = Q1 - 1.5 * iqr
sup = Q3 + 1.5 * iqr
df_filtrado = df[df['math score'].between(inf, sup)]
total_atipicos = df[(df['math score'] < inf) | (df['math score'] > sup)].shape[0]
print(total_atipicos)
```

```python
print(df[~df['math score'].between(inf, sup)])
```
